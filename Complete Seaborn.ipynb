# Matrix plot


heatmap

```python
df.head()
```

```python
pivot_table=pd.pivot_table(data=df,index='City',columns='Payment',values='cogs')
print(pivot_table)
```

```python
sns.heatmap(data=pivot_table)
plt.plot()
```

```python
sns.heatmap(data=pivot_table,vmin=200,vmax=300)
plt.plot()
```

```python
sns.heatmap(data=pivot_table,vmin=200,vmax=300,cmap='Spectral_r')
plt.plot()
```

```python
sns.heatmap(data=pivot_table,vmin=200,vmax=300,annot=True,fmt='.0f')
plt.plot()
```

```python
sns.heatmap(data=pivot_table,vmin=200,vmax=300,annot=True,fmt='.0f',annot_kws=dict(size=30,weight='bold',color='red'))
plt.plot()
```

```python
sns.heatmap(data=pivot_table,vmin=200,vmax=300,annot=True,fmt='.0f',annot_kws=dict(size=30,weight='bold',color='red'),linewidth=1.5)
plt.plot()
```

cluster map

```python
sns.clustermap(data=pivot_table,vmin=200,vmax=300,annot=True,fmt='.0f',annot_kws=dict(size=30,weight='bold',color='red'),linewidth=1.5)
plt.plot()
```

```python
sns.clustermap(data=pivot_table,vmin=200,vmax=300,annot=True,fmt='.0f',annot_kws=dict(size=30,weight='bold',color='red'),linewidth=1.5,
              col_cluster=None)
plt.plot()
```
